### Hazen estimation of K:

```python
d_10 = np.array([0.004,0.06,0.5])
K = 0.0116*d_10**2
```

```python
K
```

### Estimating Volumes with Darcy's Law

```python
dh = (115-120)*1e-2
dx = 1
D = 0.1
A = np.pi*D**2/4
Q = -K*(dh/dx)*A
answer = np.around((Q*3600*24)*1000,3)
```

```python
answer
```

```python
print(f'The volume colected for silty sand is: {answer[0]} L,\n\
The volume collected for the fine sand is: {answer[1]} L,\n\
and the volume in the gravel experiment is {answer[2]} L')
```
